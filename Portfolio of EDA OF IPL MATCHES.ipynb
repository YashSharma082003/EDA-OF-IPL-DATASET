## Exploratory Data Analysis (EDA)

#### What is EDA? Why do we do EDA?
Exploratory Data Analysis refers to the critical process of performing initial investigations on data to discover patterns, to spot anomalies, test hypotheses and check assumptions with the help of summary statistics and graphical representations.

Exploratory Data Analysis (EDA) is an approach to analysing the data using visual techniques. It is used to discover trends, and patterns, or to check assumptions with the help of statistical summaries and graphical representations.

Exploratory Data Analysis involves two fundamental steps:-
* Data Analysis(Data Preprocessing, Cleaning and Manipulation).
* Data Visualisation(Visualize relationships in data using different types of plots).

#### About IPL Dataset 
Indian Premier League(IPL) is a professional Twenty 20 cricket league in India contested during March or April and May of every year by eight teams representing eight different cities in India. The league was founded by the Board of Control for Cricket in India (BCCI) in 2008.
The data set that is going to be used is the 'IPL Matches Dataset'.It contains IPL matches data from 2008 to 2019, we are going to clean the data if required and perform Exploratory Data Analysis(EDA). The tools that are going to be used for the EDA would be NumPy, Pandas, Matplotlib and Seaborn.

Q1 Number of matches played in each IPL season.

```python
df['Season'].value_counts()
plt.figure(figsize=(12,6))
sns.countplot(x='Season',data=df,
            order=df['Season'].value_counts().index.sort_values(),
            palette='Set1')
plt.title("NUMBER OF MATCHES PLAYED IN EACH IPL SEASON",fontsize=30)
plt.xlabel("SEASON",fontsize=20)
plt.ylabel("NUMBER OF MATCHES PLAYED",fontsize=20)
plt.show()
```

##### INFERENCES
<pre>
* There were most matches played in IPL-2013.
* The Three Seasons in which the most matches were played are IPL-2011, IPL-2012 and IPL-2013.
<pre>

Q2 Toss Winner(Team winning most tosses) across all IPL seasons.

```python
df['toss_winner'].value_counts()
plt.figure(figsize=(12,6))
sns.countplot(y='toss_winner',data=df,
            order=df['toss_winner'].value_counts().index,
            palette='magma')
plt.title("TOSS WINNER ACROSS ALL IPL SEASONS",fontsize=30)
plt.xlabel("NUMBER OF TOSS WINNING",fontsize=20)
plt.ylabel("TOSS WINNING TEAM",fontsize=20)
plt.show()
```

##### INFERENCES
<pre>
* Mumbai Indians Team was winning most of the toss across all IPL Seasons.
* The Three Teams whose win most of the toss across all IPL Seasons were Mumbai Indians, Kolkata Knight Riders, Chennai Super Kings.
<pre>

Q3 Toss Decision(Bat or to Field) in each IPL seasons.

```python
df['toss_decision'].value_counts()
plt.figure(figsize=(12,6))
sns.countplot(x='Season',hue='toss_decision',data=df,
            order=df['Season'].value_counts().index,
            palette='Set2')
plt.title("TOSS DECISION IN EACH IPL SEASONS",fontsize=30)
plt.xlabel("SEASONS",fontsize=20)
plt.ylabel("NUMBER OF TOSS DECISION",fontsize=20)
plt.show()
```

##### INFERENCES
<pre>
* There were 463 decisons taken of Field.
* There were 293 decisons taken of Bat.
<pre>

Q4 Toss Decision Percentage across all IPL seasons.

```python
a = df['toss_decision'].value_counts()
plt.figure(figsize=(12,6))
plt.pie(x=a.values,labels=a.index,colors=['lightgreen','lightblue'],autopct='%.01f%%',explode=(0,0.03))
plt.title("TOSS DECISION ACROSS ALL IPL SEASONS",fontsize=20)
plt.show()
```

##### INFERENCES
<pre>
* There is 61.20% toss decision taken in all IPL Seasons.
* There is 38.80% toss decision taken in all IPL Seasons.
<pre>

Q11 Find that teams whose win the toss and win the match.

```python
df['win_the_match']=(df['toss_winner']==df['winner'])
df['win_the_match'] = np.where(df['win_the_match']==True,'Win','Loose')
```

```python
q = df['win_the_match'].value_counts()
plt.figure(figsize=(12,6))
plt.bar(q.index,q.values,color=['red','blue'])
plt.title("FIND THAT TEAMS WHOSE WIN THE TOSS AND WIN THE MATCH",fontsize=20)
plt.xlabel("WIN/LOOSE",fontsize=15)
plt.ylabel("NUMBER OF TIMES MATCHES WON",fontsize=15)
plt.show()
```

##### INFERENCES
<pre>
Across all the IPL seaosns in 393 matches, the toss winner emerged as the match winner as well and in 363 matches the toss winner lost the match.
<pre>

Q12 Win by Runs(How many times each IPL team has won the match by runs).

```python
count_win_by_runs_per_team = {}
for i in df['team1'].unique():
    win_team = df[df['winner']==i]
    count_win_by_runs_per_team[i] = win_team[win_team['win_by_runs']!=0].shape[0]
```

```python
plt.figure(figsize=(12,6))
plt.barh(list(count_win_by_runs_per_team.keys()),list(count_win_by_runs_per_team.values()))
plt.title("IPL TEAM WISE TOTAL WIN BY RUNS",fontsize=20)
plt.xlabel("COUNT OF WIN BY RUNS",fontsize=15)
plt.ylabel("IPL TEAMS",fontsize=15)
plt.grid()
plt.show()
```

##### INFERENCES
<pre>
MI has won most of the matches by winning by Run, followed by Chennai Super Kings
<pre>

Q13 Win by Wickets(How many time seach IPL team has won the match by wickets).

```python
count_win_by_wickets_per_team = {}
for i in df['team1'].unique():
    win_team = df[df['winner']==i]
    count_win_by_wickets_per_team[i] = win_team[win_team['win_by_wickets']!=0].shape[0]
```

```python
plt.figure(figsize=(12,6))
plt.barh(list(count_win_by_wickets_per_team.keys()),list(count_win_by_wickets_per_team.values()),color='green')
plt.title("IPL TEAM WISE TOTAL WIN BY WICKETS",fontsize=20)
plt.xlabel("COUNT OF WIN BY WICKETS",fontsize=15)
plt.ylabel("IPL TEAMS",fontsize=15)
plt.grid()
plt.show()
```

##### Inference
<pre> 
Kolkata Knight Riders have won most of the IPL matches based on win by wickets, followed by Mumbai Indians and Chennai Super Kings
<pre>

##### Team Wise Analsyis

### MUMBAI INDIANS

```python
mi = df[(df['team1']=='Mumbai Indians') | (df['team2']=='Mumbai Indians')]
p = mi['winner'].value_counts()
plt.figure(figsize=(12,6))
plt.barh(width=p.values,y=p.index)
plt.title('MI Win performance over other teams')
plt.ylabel('Teams')
plt.xlabel('Number of matches won')
plt.grid()
plt.show()
```

##### INFERENCES
<pre>
MI has won 109 times when compared to the losses against it.
<pre>

```python
csk = df[(df['team1']=='Chennai Super Kings') | (df['team2']=='Chennai Super Kings')]
r = csk['winner'].value_counts()
plt.figure(figsize=(12,6))
plt.barh(width=r.values,y=r.index,color='green')
plt.title('CSK Win performance over other teams')
plt.ylabel('Teams')
plt.xlabel('Number of matches won')
plt.grid()
plt.show()
```
